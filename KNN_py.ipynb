```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

```
df = pd.read_csv('sample_data/knn_test.csv')
df.head()
```

```
plt.scatter(df['x'], df['y'], c=df['classe'])
plt.show()
```

```
dado_treino,dado_teste, grupo_treino, grupo_teste = train_test_split(df[
      ['x','y']].values,df
      ['classe'].values,test_size=0.3)

print('%d exemplos de treinamento' % len(dado_treino))
print('%d exemplos de treinamento' % len(dado_teste))
```

```
modelo = KNeighborsClassifier(n_neighbors=3)
modelo.fit(dado_treino, grupo_treino)
acerto = modelo.score(dado_teste,grupo_teste)
acerto = acerto * 100
print('Percentual de acerto: %.4f' % acerto)
```

```
x= 100
print ('Digite 99 no valor para sair')
while(x!=99):
  x = float(input('Informe ponto no eixo x: \n'))
  if(x!=99):
    y = float(input('Informe ponto no eixo y: \n'))

    ponto_a_ser_predito = [[x,y]]
    predicao = modelo.predict(ponto_a_ser_predito)
    if (predicao==0):
      print('ROXO')
    else:
      print('AMARELO')
```
